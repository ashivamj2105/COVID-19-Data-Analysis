```python
df_covid.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace=True, axis =1)
```

```python
df_covid['Date'] = pd.to_datetime(df_covid['Date'], format= '%Y-%m-%d')
df_covid.head(5)
```

```python
df_covid['Active_cases'] = df_covid['Confirmed'] - (df_covid['Cured'] + df_covid['Deaths'])
df_covid.tail()
```

```python
statewise = pd.pivot_table(df_covid, values= ["Confirmed", "Deaths", "Cured"], index= "State/UnionTerritory", aggfunc = max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"] * 100/ statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"] * 100/ statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by= "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
top_10_active_states = df_covid.groupby(by = "State/UnionTerritory").max()[['Active_cases','Date']].sort_values(by = ['Active_cases'], ascending=False).reset_index()
fig = plt.figure(figsize = (16, 9))
plt.title("Top 10 states with most active cases in India", size = 18)
ax = sns.barplot(data = top_10_active_states.iloc[:10], y = "Active_cases", x = "State/UnionTerritory" , linewidth = 2, edgecolor = 'black')
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
top_10_death_states = df_covid.groupby(by= 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
fig = plt.figure(figsize= (18, 6))
plt.title("Top 10 states with most deaths in India", size = 18)
ax = sns.barplot(data = top_10_death_states.iloc[:12], y = 'Deaths', x = 'State/UnionTerritory', linewidth = 2, edgecolor = 'black')
plt.xlabel("States")
plt.ylabel("Deaths")
plt.show()
```

```python
fig = plt.figure(figsize = (12,6))
ax = sns.lineplot(x = 'Date',y = 'Active_cases', hue = 'State/UnionTerritory',data = df_covid[df_covid['State/UnionTerritory'].isin(['Maharastra', 'Karnataka','Kerala', 'Tamil Nadu', 'Uttar Pradesh'])])
ax.set_title("Top 5 Affected States in India", size = 15)
```

```python
vaccine = df_vaccine[df_vaccine.State!= 'India']
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaccinated": "Total"}, inplace = True)
vaccine.head()
```

```python
most_vaccinated_state = vaccine.groupby('State')['Total'].sum().to_frame('Total')
most_vaccinated_state = most_vaccinated_state.sort_values('Total', ascending = False)[:5]
most_vaccinated_state
```

```python
fig = plt.figure(figsize = (10, 5))
plt.title("Top 5 vaccinated states in India", size = 20)
x = sns.barplot(data = most_vaccinated_state.iloc[:10], y = most_vaccinated_state.Total, x = most_vaccinated_state.index, linewidth = 2, edgecolor = 'black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
